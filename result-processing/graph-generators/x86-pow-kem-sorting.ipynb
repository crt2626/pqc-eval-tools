```python
"""Imports"""
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
from io import StringIO
```

```python
"""Creating the x86 Powerful Graphs for Iterations - Iterations All"""
#Declaring lists
alg_kem_list_unfilt = []
kem_list_operation_unfilt = []
alg_kem_list = []
kem_list_operation = []

#Importing ARM Iteration
df_arm_og = pd.read_csv("../csv-files/x86-pow-results-kem-filt-prop.csv", index_col=False)
df_arm_iteration = df_arm_og[["Algorithm","Operation","Iterations"]]

#Assigning list values and removing trailing whitespace
alg_kem_list_unfilt = df_arm_iteration['Algorithm'].unique().tolist()
kem_list_operation_unfilt = df_arm_iteration['Operation'].tolist()
kem_list_iteration = df_arm_iteration['Iterations'].tolist()

for list_item in alg_kem_list_unfilt:
    moded_list_item = list_item.split(" ")
    alg_kem_list.append(moded_list_item[0])

for list_item in kem_list_operation_unfilt:
    moded_list_item = list_item.split(" ")
    kem_list_operation.append(moded_list_item[0])

#Setting bar parameters
nump = np.arange(len(alg_kem_list))
width = 0.25
plt.barh(nump-(width*2), kem_list_iteration[0::3], width, label='Key-Generation', zorder=2)
plt.barh(nump-width, kem_list_iteration[1::3], width, label='Encapsulation', zorder=2)
plt.barh(nump, kem_list_iteration[2::3], width, label='Decapsulation', zorder=2)
plt.yticks(nump - 0.25, alg_kem_list)

#Sytling the graph
plt.rcParams["figure.autolayout"] = True
plt.rcParams["figure.figsize"] = [10, 10]
plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
plt.grid(axis='x', linestyle='--', alpha=0.7, zorder=0)

#Adding titles and plotting
plt.xlabel("Number of Iterations", labelpad=15, fontweight='bold')
plt.ylabel("Algorithm Name", labelpad=25, fontweight='bold')
plt.title("x86 Powerful CPU KEM Benchmarking - Iterations", fontweight='bold')
plt.legend()
plt.show()
```
